```python
#importing needed library
import pandas as pd
import numpy as np
from sklearn.preprocessing import OneHotEncoder
```

```python
#Creating dataset
data = {'Temperature': ['Very Hot','Hot', 'Cold', 'Very Hot', 'Warm', 'Hot', 'Warm', 'Hot', 'Cold']}
df = pd.DataFrame(data, columns = ['Temperature'])
df
```

```python
# One hot Encoding
ohc = OneHotEncoder()
df_new = pd.DataFrame(ohc.fit_transform(df[['Temperature']].values.reshape(-1,1)).toarray())
df_ohc = pd.concat([df, df_new], axis=1)
df_ohc
```

```python
#Dummy Encoding 
df_dummy = pd.get_dummies(df, prefix = ['Temp'], columns = ['Temperature'])
df_dummy = pd.concat([df, df_dummy], axis=1)
df_dummy
```

```python
df_dummy_t = pd.get_dummies(df, prefix = ['Temp'], columns = ['Temperature'], drop_first=True)
df_dummy_t = pd.concat([df, df_dummy_t], axis=1)
df_dummy_t
```
