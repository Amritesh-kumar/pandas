```python
#importing needed library
import pandas as pd
import numpy as np
from sklearn.preprocessing import LabelEncoder
```

```python
#Creating dataset
data = {'Temperature': ['Very Hot','Hot', 'Cold', 'Very Hot', 'Warm', 'Hot', 'Warm', 'Hot', 'Cold']}
df = pd.DataFrame(data, columns = ['Temperature'])
df
```

```python
#using LabelEncoder 
df['Temp_label_encoded'] = LabelEncoder().fit_transform(df['Temperature'])
df
```

Cold<Hot<Very Hot<Warm   -----  0 < 1 < 2 < 3 

```python
#using map function
df['Temperature_maped'] = df['Temperature'].map({'Cold':0, 'Hot':1, 'Very Hot':2, 'Warm':3}).astype(np.int)
df
```
